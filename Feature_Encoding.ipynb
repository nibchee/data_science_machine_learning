```python
### Label Encoding ###
```

```python
le=preprocessing.LabelEncoder()
df['Gender_label']=le.fit_transform(df['Gender'])
df.head(5)
```

```python
### One Hot Encoding Technique ###
```

```python
one_hot=pd.get_dummies(df['Geography'])
one_hot.head(5)
```
