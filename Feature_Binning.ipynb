```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df=pd.read_csv('/content/sample_data/Churn_Modelling.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop(columns=['CustomerId','RowNumber','Surname'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df.Age.min()
```

```python
df.Age.max()
```

```python
labels=['0-20','21-40','41-60','Above 61']
bins=[0,20,40,60,100]
df['Age_bins']=pd.cut(df.Age,bins,labels=labels,include_lowest=True)
```

```python
df.head(5)
```

```python
df[['Age','Age_bins']].to_csv('Age_bins.csv')
```

df.Age_bins.value_counts()

```python
#making bar chart on data
plt.bar(labels,df.Age_bins.value_counts())
plt.xlabel('Age_bins')
plt.ylabel('Count')
plt.show()
```
