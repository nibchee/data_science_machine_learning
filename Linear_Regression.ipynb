```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/50_Startups.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Divide Data into depenedents & Independents
X=df.iloc[:,:-1].values
y=df['Profit'].values
```

```python
X
```

```python
y
```

```python
#Feature Encoding- OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import numpy as np
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
X=ct.fit_transform(X)
```

```python
X
```

```python
#Train and Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#Create LR Model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
#Predictions
y_pred=lr.predict(X_test)
```

```python
#Plotting Y_pred / y_test
plt.plot(y_test,color='blue',label='test')
plt.plot(y_pred,color='red',label='pred')
plt.show()
```

```python
#Out Of Box Predictions
data=[[0.0,0.0,1.0,16000,130000,350000]]
new_df=pd.DataFrame(data)
```

```python
new_df=sc.transform(new_df)
```
