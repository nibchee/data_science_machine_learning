```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
telco_base_data=pd.read_csv('/content/sample_data/Churn_Modelling.csv')
```

```python
telco_base_data.head(5)
```

```python
telco_new=telco_base_data[['Geography','Gender','Exited']]
```

```python
telco_new.head(5)
```

```python
for i, predictor in enumerate(telco_new.drop(columns=['Exited'])):
  plt.figure()
  sns.countplot(data=telco_new,x=predictor,hue='Exited')
```

```python
sns.histplot(x='Gender', hue='Geography', date=telco_new, stat="count")
```

```python
telco_base_data.select_dtypes(include=np.number).corr()['Exited'].sort_values(ascending=False).plot(kind='bar')
```
